```python
# Import loading optimization class under policy 1b
from LoadingOptimizationPolicy1b import LoadingOptimizationPolicy1b

# Load sample data for testing
import SampleData as sample_data
```

```python
# From the SampleData class get necessary data for the inputs of the LoadingOptimizationPolicy1b
am_types, route, slot_ids = sample_data.get_sample_data()
constraints = sample_data.get_sample_constraints()
```

```python
# Initialize the optimization class with necessary inputs
optimizer = LoadingOptimizationPolicy1b(
    route=route,
    constraints=constraints,
    am_types=am_types,
    slot_ids=slot_ids,
    verbose=False)

# Solve
optimizer.run()
# Get solution
solution = optimizer.get_solution()
```

```python
print("Loading cost:", solution["count_reloads"])
print("Solution time:", solution["solution_time"])
print("Loading plan:", solution["loading_plan"])
```

```python
# Draw the Space State network
optimizer.draw()
```
